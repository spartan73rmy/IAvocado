### Load Model

```python
model = tf.keras.models.load_model('saved_model/model2')

# Check its architecture
#model.summary()
```

```python
import pathlib

data_dir=r"C:\Users\spart\Desktop\Source\IAvocado\testing" #rename images plath
data_dir = pathlib.Path(data_dir)
```

```python
images=list(data_dir.glob('*/*.jpg'));
print(len(images))
```

```python
arr=[];
plt.figure(figsize=(10, 10))

for i in images:
    image = tf.keras.preprocessing.image.load_img(str(i),target_size = (img_height,img_width))
    input_arr = keras.preprocessing.image.img_to_array(image)    
    input_arr = np.array([input_arr])  # Convert single image to a batch.
    arr.append(input_arr);
    predictions = model.predict(input_arr)
    print(predictions)
    imgplot = plt.imshow(image)
    plt.show()
```
